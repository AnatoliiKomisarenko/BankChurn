```python
# Разбиваем данные на группы по признаку оттока (churn)
df_churn = df[df['churn'] == 1]
df_nchurn = df[df['churn'] == 0]

# Функция для визуальной оценки распределения и Q-Q plot
def plot_distribution(data, column, title_suffix=''):
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    sns.histplot(data[column], kde=True, ax=axes[0])
    axes[0].set_title(f'Гистограмма {column} {title_suffix}')
    stats.probplot(data[column], dist="norm", plot=axes[1])
    axes[1].set_title(f'Q-Q plot {column} {title_suffix}')
    plt.show()
```

```python
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = max(0, Q1 - 1.5 * IQR)
    upper_bound = Q3 + 1.5 * IQR

    print(f'Границы IQR для {column}: ниже {lower_bound}, выше {upper_bound}')
    
    # Убираем выбросы
    filtered_data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    print(f'Удалено выбросов для {column}: {len(data) - len(filtered_data)}')
    return filtered_data
```

Критерий Манна-Уитни для переменной products
* Обоснование:
    * Критерий Манна-Уитни (U-тест) используется для сравнения двух независимых групп (например, churn = 0 и churn = 1) в случае, если данные:

*   Предпосылки для применения U-теста и их проверка:

    * Природа данных:
        * Переменная products принимает дискретные значения (целые числа >= 0), что не нарушает требования для U-теста. При этом данные не обязаны быть непрерывными.
        * Нормальность распределения не требуется:
            * U-тест является непараметрическим, поэтому отсутствие нормальности в данных не влияет на его достоверность.
        * Независимость выборок:
            * Группы данных (churn = 0 и churn = 1) независимы. Это предпосылка соблюдается, если каждая строка соответствует уникальному клиенту.
        * Размер выборок:
            * U-тест применим к данным любых размеров, но чем больше выборка, тем выше мощность теста.

```python
u_stat_prod, u_p_prod = stats.mannwhitneyu(df_churn['products'], df_nchurn['products'], alternative='two-sided')
print("\nМанна‑Уитни U‑тест для products:")
print(f"U-статистика = {u_stat_prod:.3f}, p-value = {u_p_prod:.3f}")


if u_p_prod < 0.05:
    # Обычно, если у оставшихся клиентов больше продуктов, то среднее значение будет выше
    if df_nchurn['products'].mean() > df_churn['products'].mean():
        print("Результат: Клиенты с большим количеством продуктов (non-churn) имеют статистически существенно большее их число.")
    else:
        print("Результат: Напротив, клиенты, ушедшие из банка, имеют больше продуктов – гипотеза требует уточнения.")
else:
    print("Результат: Нет статистически значимых различий между группами по количеству продуктов.")
```

Хи-квадрат тест для переменных credit_card
* Обоснование:
    * Хи-квадрат тест используется для проверки независимости двух категориальных переменных. В данном случае требуется определить, связана ли переменная credit_card (наличие/отсутствие кредитной карты) с переменной churn (отток клиента).

* Предпосылки для применения Хи-квадрат теста и их проверка:

    * Категориальные переменные:
        * Переменная credit_card является бинарной (1 = наличие карты, 0 = отсутствие). Аналогично, переменная churn бинарна (1 = клиент ушел, 0 = клиент остался), что делает эти данные подходящими для хи-квадрат теста.
    * Наблюдения независимы:
        * Клиенты в обеих категориях (churn = 0 и churn = 1) независимы друг от друга. Предполагается, что каждой строке данных соответствует уникальный клиент, что делает выборки независимыми.
    * Ожидаемые частоты в таблице сопряженности:
        * В каждой ячейке таблицы сопряженности (например, credit_card = 1 и churn = 1) ожидаемое количество наблюдений должно быть >= 5.
    * Размер выборки:
        * Тест применим на больших выборках, но в малых выборках может быть ненадежен.

```python
# Создадим кросс-таблицу между churn и credit_card
ct = pd.crosstab(df['churn'], df['credit_card'])
print("\nКросс-таблица для переменной credit_card и churn:")
print(ct)

# Применим хи‑квадрат тест для проверки взаимосвязи
chi2, chi2_p, dof, expected = stats.chi2_contingency(ct)
print("\nХи‑квадрат тест для credit_card:")
print(f"chi2 = {chi2:.3f}, p-value = {chi2_p:.3f}, степени свободы = {dof}")

if chi2_p < 0.05:
    print("Результат: Наличие кредитной карты статистически связано с оттоком.")
else:
    print("Результат: Нет статистически значимой связи между наличием кредитной карты и оттоком.")
```

### Шаг 5 Кластеризация клиентов

```python
df.info()
```

```python
df = pd.get_dummies(df, columns=['gender'], prefix='gender', drop_first=True)
```

```python
features = ['balance', 'last_activity', 'products', 'score', 'age', 'equity', 'est_salary', 'gender_М', 'credit_card']
X = df[features].copy()
```

```python
X.info()
```

```python
X = X.dropna()
```

```python
df = df.loc[X.index]
```

```python
# Нормализация данных
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Определение оптимального числа кластеров методом локтя
sse = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(k_range, sse, marker='o')
plt.xlabel('Число кластеров (k)')
plt.ylabel('SSE (сумма квадратов отклонений)')
plt.title('Метод локтя для определения оптимального k')
plt.show()
```

```python
# Кроме того, можно рассчитать коэффициент силуэта для выбора лучшего k
silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, cluster_labels)
    silhouette_scores.append(score)
    
plt.figure(figsize=(8, 4))
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.xlabel('Число кластеров (k)')
plt.ylabel('Коэффициент силуэта')
plt.title('Определение оптимального k по коэффициенту силуэта')
plt.show()
```

```python
# После анализа графиков выбираем, например, k = 4 (учитывая, что кластеры должны быть не менее 500 клиентов)
optimal_k = 4

# Проведем кластеризацию с использованием KMeans
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['cluster'] = kmeans.fit_predict(X_scaled)
```

```python
# Проверка размеров кластеров
cluster_sizes = df['cluster'].value_counts().sort_index()
print("Размеры кластеров:")
display(cluster_sizes)
```

```python
# Анализ параметров и уровня оттока по кластерам
cluster_summary = df.groupby('cluster').agg(
    clients=('user_id', 'count'),
    avg_score=('score', 'mean'),
    avg_products=('products', 'mean'),
    avg_age=('age', 'mean'),
    avg_equity=('equity', 'mean'),
    avg_balance=('balance', 'mean'),
    avg_est_salary=('est_salary', 'mean'),
    churn_rate=('churn', lambda x: np.mean(x) * 100)
).reset_index()
```

```python
print("\nСводная статистика по кластерам:")
display(cluster_summary)
```

* Кластер 1 имеет самый высокий отток (~60%).

```python
# Визуализация распределения оттока по кластерам
plt.figure(figsize=(8, 5))
sns.barplot(x='cluster', y='churn_rate', hue='cluster', data=cluster_summary, palette='viridis', dodge=False)
plt.xlabel('Кластер')
plt.ylabel('Уровень оттока (%)')
plt.title('Уровень оттока по кластерам')
plt.show()
```

```python
# Вывод процентного состава каждого кластера
total_clients = df.shape[0]
cluster_summary['clients_percent'] = cluster_summary['clients'] / total_clients * 100
print("\nПроцентное распределение клиентов по кластерам:")
display(cluster_summary[['cluster', 'clients', 'clients_percent']])
```

* Самый высокий отток (31%) имеет Кластер 0

```python
# Фильтрация сегментов для дальнейшей работы 
# Задаем условие: сегмент должен содержать не менее 500 клиентов и иметь уровень оттока выше среднебанковского.
# Считаем среднебанковский уровень оттока:
overall_churn_rate = df['churn'].mean() * 100
print("\nОбщий уровень оттока:", overall_churn_rate, "%")
```
